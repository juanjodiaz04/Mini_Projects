```python
import numpy as np
import matplotlib.pyplot as plt

#Radios Y Periodos Orbitales de la Tierra y Venus en UA y das
re, rv, Te, Tv = 1, 0.72, 365.25, 224.7

# Numero de aos para grficar en la tierra y posiciones 
n, N = 9, 750

# Cuadrcula y posiciones angulares 
t = np.linspace(0, n*Te, N)
thetat = t * 2*np.pi / Te
thetav = t * 2*np.pi / Tv

# Conversin de Polares a Cartesianas
xe, ye = re * np.cos(thetat), re * np.sin(thetat)
xv, yv = rv * np.cos(thetav), rv * np.sin(thetav)

#Grfico
plt.figure(figsize=(10,10))
for i in range(len(thetat)):
    plt.plot([xe[i], xv[i]], [ye[i], yv[i]], color='blue', alpha=0.3)
plt.title('Dance of the planets')
plt.axis('equal')
plt.axis('on')
plt.show()
```
